1. $r_x,_y$ - показатель силы и направления взаимосвязи двух количественных переменных

3. Коэффициент корреляции может принимать значения на промежутке [-1;1]

Гистограмма с ярко выраженным пиком (1 мода) называется унимодальной. Частным примером является нормальное распределение.

Если мы можем различить у гистограммы два ярко выраженных пика, то гистограмма называется бимодальной. Во многих случаях это значит, что выборки происходят из двух разных популяций, так как наличие двух мод в одной популяции маловероятное явление. 


Гистограммы с большим количеством пиков (многомодальные) встречаются крайне редко и, зачастую свидетельствуют о присутствии специальных факторов, влияющих на исследуемую систему или процесс. Если каждый интервал гистограммы содержит примерно равное количество значений, то такая гистограмма называется однородной или гистограммой равномерного распределения.

**ТЕОРИЯ ВЕРОЯТНОСТИ**

* $\mathbb N$ — множество натуральных чисел: 1, 2, 3, ....

* $\mathbb Z$ — множество целых чисел.

* $\mathbb Q$ — множество рациональных чисел, то есть дробей вида $\frac{m}{n}$
, где $m\in\mathbb Zm∈Z и n\in\mathbb Nn∈N$.

* $\mathbb R$ — множество вещественных (действительных) чисел.

* $\mathbb C$ — множество комплексных чисел.

* Запись $\mathbb R\times\mathbb R$ означает множество всех пар вещественных чисел.

* Записи $A\cup B  и A\cap B$ обозначают соответственно объединение и пересечение множеств AA и BB.

* Записи $A\setminus B$ обозначают разность множеств, т.е. $A\setminus B$ состоит из таких элементов множества AA, которые не принадлежат множеству BB.

* Запись $A\subset B$ означает, что AA является подмножеством BB, т.е. любой элемент из AA принадлежит BB.

* Записи $\bigcup\limits_{k=1}^n A_k$ и $\bigcap\limits_{k=1}^n A_k$
обозначают соответственно объединение и пересечение всех множеств $A_1,A_2, \dots,A_nA$.

* Записи $\bigcup\limits_{k=1}^\infty A_k$
и $\bigcap\limits_{k=1}^\infty A_k$
обозначают соответственно объединение и пересечение всех множеств $A_1,A_2, \dots,A_k,\dots A $.

* $\forall$ — квантор всеобщности, используется для сокращенной записи вместо слов “каждый“, “любой”, или “для всякого“, “для любого“ и т. п..

* ∃ — квантор существования, используется для сокращенной записи вместо слов “найдется“, “существует“ и т. п..

*Условной вероятностью* события **А** при условии, что произошло событие **В**, называется число:

$$P(A | B) = \frac{P(A \cap B)}{P(B)}$$
